```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
num_classes = 4
y_onehot = np.eye(num_classes)[y]
```

```python
def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))  # stabil
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)
```

```python
def cross_entropy(y_true, y_pred):
    m = y_true.shape[0]
    return -np.sum(y_true * np.log(y_pred + 1e-9)) / m
```

```python
np.random.seed(42)

input_size = 2
output_size = 4

W = np.random.randn(input_size, output_size) * 0.01
b = np.zeros((1, output_size))
```

```python
learning_rate = 0.01
epochs = 1000

for i in range(epochs):
    # Forward
    z = np.dot(X, W) + b
    y_pred = softmax(z)

    # Loss
    loss = cross_entropy(y_onehot, y_pred)

    # Backprop
    m = X.shape[0]
    dz = (y_pred - y_onehot) / m

    dW = np.dot(X.T, dz)
    db = np.sum(dz, axis=0, keepdims=True)

    # Update (SGD)
    W -= learning_rate * dW
    b -= learning_rate * db

    if i % 100 == 0:
        print(f"Epoch {i}, Loss: {loss}")
```

```python
# Grid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)

grid = np.c_[xx.ravel(), yy.ravel()]

# Prediksi
Z = softmax(np.dot(grid, W) + b)
Z = np.argmax(Z, axis=1)
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.xlabel("RAM (scaled)")
plt.ylabel("Battery Power (scaled)")
plt.title("Decision Boundary Linear Classifier")
plt.show()
```
